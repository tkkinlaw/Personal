```python
txt = open(r"/Users/timothy/Documents/GitHub/Personal/Advent of Code/AOC2023/Data/Day4.txt")
input = txt.read()
```

```python
cards = input.split('\n')
cardNums = []
cardWinValues = []
cardActualValues = []
for card in cards:
    print(card)
    cardNums.append(card.split(':')[0])
    cardWinValues.append(card.split(' | ')[0].split(':')[1])
    cardActualValues.append(card.split(' | ')[1])
print(cardNums)
print(cardWinValues)
print(cardActualValues)
    
```

```python
df = pd.DataFrame({'Winning Numbers':cardWinValues,
                   'Actual Values':cardActualValues}, index=cardNums)
df
```

```python
def makeIntegerLists(arg1):
    arg1.to_list()[0].replace(' ', '')
    arg1.to_list()[1].replace(' ', '')
    winValues = []
    actualValues = []
    for num in arg1[0].split(' '):
        try: 
            winValues.append(int(num))
        except:
            #winValues.append(num) # need to keep the empty strings
            pass
    for num in arg1[1].split(' '):
        try: 
            actualValues.append(int(num))
        except:
            #actualValues.append(num)
            pass
    return {"Wins":winValues,"Actuals":actualValues}
```

```python
def createScore(arg1):
    if arg1 == 1:
        score = 1
    elif arg1 == 0:
        score = 0
    else:
        score = 1
        for x in range(1, arg1):
            score = (score*2)
    return score
```

```python
cardsPoints = []
for (colName, colValue) in df.iterrows():
    intValues = makeIntegerLists(colValue)
    pairs = set(intValues['Wins']).intersection(set(intValues['Actuals']))
    cardPoint = createScore(len(list(pairs)))
    cardsPoints.append(cardPoint)
print(sum(cardsPoints))
```
