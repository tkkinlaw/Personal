## Part 1

```python
txt = open(r"/Users/timothy/Documents/GitHub/Personal/Advent of Code/AOC2023/day1Part2.txt")
input = txt.read()
```

```python
def canBeInt(arg1):
    try:
        intArg = int(arg1)
        return intArg
    except:
        pass
```

```python
calibValueSum = 0
for badLine in input.split('\n'):
    lineInts = []
    for char in badLine:
        res = canBeInt(char)
        if type(res) == type(int()):
            lineInts.append(char)
        else:
            pass
    if len(lineInts)==1:
        lineInts.append(lineInts[0])
    if len(lineInts) > 2:
        lineInts = [lineInts[0], lineInts[-1]]
    print(lineInts)
    calibValue = int(str(lineInts[0])+str(lineInts[1]))
    print(calibValue)
    calibValueSum += calibValue
print(f"The total calibration value is: {calibValueSum}")
```

## Part 2

```python
spelledNums = {
    'one':'1',
    'two':'2',
    'three':'3',
    'four':'4',
    'five':'5',
    'six':'6',
    'seven':'7',
    'eight':'8',
    'nine':'9'
}
```

```python
def findEndNum(argTxt2):
    print(argTxt2)
    for x in range(1,len(argTxt2)+1,1):
        for key in spelledNums.keys():
            if key in argTxt2[-x:]:
                argTxt2 = argTxt2.replace(key, spelledNums[key])
    return argTxt2

def convertSpelledNum(argTxt):
    for i in range(1,len(argTxt)+1,1):
        print(argTxt[0:i])
        for key in spelledNums.keys():
            if key in argTxt[0:i]:
                argTxt = argTxt.replace(key, spelledNums[key])
    return argTxt
test = "eightwothree"
res2 = convertSpelledNum(test) + findEndNum(test)
print(res2)
```

```python
listInts = []

for badLine in input.split('\n'):
    print(badLine)
    badLine = convertSpelledNum(badLine) + findEndNum(badLine)
    print(badLine)
    lineInts = []
    for char in badLine:
        res = canBeInt(char)
        if type(res) == type(int()):
            lineInts.append(char)
        else:
            pass
    if len(lineInts)==1:
        lineInts.append(lineInts[0])
    if len(lineInts) > 2:
        lineInts = [lineInts[0], lineInts[-1]]
    print(lineInts)
    calibValue = int(str(lineInts[0])+str(lineInts[-1]))
    print(calibValue)
    listInts.append(calibValue)
print(f"The total calibration value is: {sum(listInts)}")
```
