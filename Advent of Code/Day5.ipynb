```python
startCargo = data.split("\n\n")[0]
craneMoves = data.split("\n\n")[1].split('\n')

colspecs = [(1,2), (5,6), (9,10), (13,14), (17,18), (21,22), (25,26), (29,30), (33,34)]
names = []
i=1
for col in colspecs:
    names.append("stack"+str(i))
    i+=1

write_file = open(r"C:\Users\tkkin\Downloads\day5_crates.txt", 'w')
write_file.write(startCargo)
write_file.close()
read_file = open(r"C:\Users\tkkin\Downloads\day5_crates.txt", 'r')
dfCrates = pd.read_fwf(read_file, colspecs=colspecs, names=names)
dfCrates
```

```python
class stack:
    def __init__(self, inDF, stackNum):
        self.inDF = inDF
        self.stackNum = stackNum
    def clearNulls(self):
        self.inDF = self.inDF.replace(np.nan, '', regex=True)
        self.newList = self.inDF['stack'+str(self.stackNum)].tolist()
        while('' in self.newList):
            self.newList.remove('')
            #self.newList.reverse()
        return self.newList
```

```python
df_craneMoves = pd.DataFrame({
    "Instruction": craneMoves,
    "Amount": None,
    "Origin": None,
    "Destination": None
})
```

```python
i = 0
for move in craneMoves:
    df_craneMoves['Amount'][i] = move[5]
    df_craneMoves['Origin'][i] = move[12]
    df_craneMoves['Destination'][i] = move[17]
    i += 1
print(df_craneMoves)
```

```python
stack1 = stack(dfCrates,1).clearNulls()
stack2 = stack(dfCrates,2).clearNulls()
stack3 = stack(dfCrates,3).clearNulls()
stack4 = stack(dfCrates,4).clearNulls()
stack5 = stack(dfCrates,5).clearNulls()
stack6 = stack(dfCrates,6).clearNulls()
stack7 = stack(dfCrates,7).clearNulls()
stack8 = stack(dfCrates,8).clearNulls()
stack9 = stack(dfCrates,9).clearNulls()
stackDict = {
    'stack1':stack1,
    'stack2':stack2,
    'stack3':stack3,
    'stack4':stack4,
    'stack5':stack5,
    'stack6':stack6,
    'stack7':stack7,
    'stack8':stack8,
    'stack9':stack9
}
for move in craneMoves:
    amount = move.split(" ")[1]
    origin = move.split(" ")[3]
    destination = move.split(" ")[5]
    print("Move "+ amount + " from " + origin + " to " + destination)

    destStack = stackDict['stack'+str(destination)]
    originStack = stackDict['stack'+str(origin)]
    inComingCrates = stackDict['stack'+str(origin)][0:int(amount)]

    #Move content from origin to destination
    inComingCrates.reverse()

    stackDict['stack'+str(destination)] = inComingCrates+destStack
    stackDict['stack'+str(origin)] = originStack[int(amount):]
```

```python
result = ""
for l in stackDict.values():
    result += l[0]
print(result)
```
