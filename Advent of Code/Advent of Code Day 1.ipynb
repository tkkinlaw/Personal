```python
count = elfCalText.count('\n\n')

elfCalDict = {}
for i in range(count + 1):
    if i == 0:
        pass
    else:
        keyName = "Elf" + str(i)
        elfCalDict[keyName] = []
elfCalDict
```

```python
elfCal_string = elfCalText.replace("\n", ",").split(",")

i=1

for entry in elfCal_string:
    key = "Elf"
    if i > count:
        break
    elif entry != '':
        elfCalDict[key+str(i)].append(entry)
    elif entry == '':
        i+=1
elfCalDict
```

```python
totalCalDict = {}
elfName = []
calLoad = []

for key in zip(elfCalDict.keys(), elfCalDict.values()):
    intList = list(map(int, key[1]))
    totalCal = sum(intList)
    print(str(key[0] + " has " + str(totalCal) + " calories"))
    
    elfName.append(key[0])
    calLoad.append(totalCal)
```

```python
import pandas as pd
df = pd.DataFrame(calLoad, index = elfName, columns = ["Calorie Total"])
df.sort_values(by="Calorie Total", ascending=False)
```
